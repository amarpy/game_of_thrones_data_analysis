```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,15)
```

```python
data = pd.read_csv("got-actor-screentime.csv")
print(type(data), data.shape, sep = "\n")
```

```python
data.head(10)
```

```python
grouped_df = data.groupby("actor", sort = False).sum()
grouped_df.head(25)
```

```python
grouped_df.tail(5)
```

```python
grouped_df.describe()
```

```python
grouped_df_top_20 = grouped_df.head(20)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.suptitle("Jon Snow got the highest screen time for all seasons. Tyrion close behind. \n Ned Stark got the highest screen time for a single season. ", fontsize=16)

grouped_df_top_20.plot(kind = "bar",  stacked = True, fontsize = 13, ax = ax)
```

```python
grouped_df_top_10 = grouped_df.head(10)

fig, ax = plt.subplots(nrows=4, ncols=2)
fig.suptitle("Season-wise screen time of actors", fontsize=16)

plt.subplot(421)
grouped_df_top_10["season 1"].sort_values(ascending = True).plot(kind = "barh", title = "season 1")

plt.subplot(422)
grouped_df_top_10["season 2"].sort_values(ascending = True).plot(kind = "barh", title = "season 2")

plt.subplot(423)
grouped_df_top_10["season 3"].sort_values(ascending = True).plot(kind = "barh", title = "season 3")

plt.subplot(424)
grouped_df_top_10["season 4"].sort_values(ascending = True).plot(kind = "barh", title = "season 4")

plt.subplot(425)
grouped_df_top_10["season 5"].sort_values(ascending = True).plot(kind = "barh", title = "season 5")

plt.subplot(426)
grouped_df_top_10["season 6"].sort_values(ascending = True).plot(kind = "barh", title = "season 6")

plt.subplot(427)
grouped_df_top_10["season 7"].sort_values(ascending = True).plot(kind = "barh", title = "season 7")

plt.show()
```

```python
fig, ax = plt.subplots(nrows=5, ncols=2)
fig.suptitle("Trends in Screen Time of Top 8 Actors over Over 7 Seasons", fontsize=16)

plt.subplot(421)
grouped_df.loc["Jon Snow"].plot(kind = 'line', title = "Jon Snow", fontsize = 11, x_compat=True)

plt.subplot(422)
grouped_df.loc["Tyrion Lannister"].plot(kind = 'line', title = "Tyrion Lannister", fontsize = 11, x_compat=True)

plt.subplot(423)
grouped_df.loc["Daenerys Targaryen"].plot(kind = 'line', title = "Daenerys Targaryen", fontsize = 11, x_compat=True)

plt.subplot(424)
grouped_df.loc["Sansa Stark"].plot(kind = 'line', title = "Sansa Stark", fontsize = 11, x_compat=True)

plt.subplot(425)
grouped_df.loc["Cersei Lannister"].plot(kind = 'line', title = "Cersei Lannister", fontsize = 11, x_compat=True)

plt.subplot(426)
grouped_df.loc["Arya Stark"].plot(kind = 'line', title = "Arya Stark", fontsize = 11, x_compat=True)

plt.subplot(427)
grouped_df.loc["Jaime Lannister"].plot(kind = 'line', title = "Jaime Lannister", fontsize = 11, x_compat=True)

plt.subplot(428)
grouped_df.loc["Samwell Tarly"].plot(kind = 'line', title = "Samwell Tarly", fontsize = 11, x_compat=True)


plt.show()
```

```python
ax = plt.axes()
ax.set_title("Heatmap showing the Variations in Screen Time of 20 Prominent Actors for 7 Seasons\n", fontsize = 16)
sns.heatmap(grouped_df_top_20, annot=True, fmt = ".2f",  linewidth = 1, 
            linecolor = 'k', robust = True, cmap = 'Reds', ax = ax)
```
