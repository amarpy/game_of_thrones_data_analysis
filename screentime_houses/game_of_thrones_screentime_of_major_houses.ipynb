```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,15)
```

```python
data = pd.read_csv("got-houses-screentime.csv")
print(type(data), data.shape, sep = "\n")
```

```python
data.head(5)
```

```python
grouped_df = data.groupby("house", sort = False).sum()
grouped_df
```

```python
grouped_df.describe()
```

```python
houses = grouped_df.index.tolist()
seasons = list(grouped_df.columns.values)
print(houses, seasons, sep = "\n")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.suptitle("Starks Got the Highest Screen Time, Followed by Lannisters", fontsize=16)

grouped_df.sort_values(by = ["season 1", "season 2", "season 3", "season 4", 
                             "season 5", "season 6", "season 7"], 
                       ascending = True).plot(kind = "barh", stacked = True, fontsize = 13, ax =ax)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
fig.suptitle("Starks Got the Highest Screen Time, Followed by Lannisters", fontsize=16)


plt.subplot(331)
grouped_df["season 1"].sort_values(ascending = True).plot(kind = "barh", title = "season 1",fontsize = 11)

plt.subplot(332)
grouped_df["season 2"].sort_values(ascending = True).plot(kind = "barh", title = "season 2", fontsize = 11)

plt.subplot(333)
grouped_df["season 3"].sort_values(ascending = True).plot(kind = "barh", title = "season 3", fontsize = 11)

plt.subplot(334)
grouped_df["season 4"].sort_values(ascending = True).plot(kind = "barh", title = "season 4", fontsize = 11)

plt.subplot(335)
grouped_df["season 5"].sort_values(ascending = True).plot(kind = "barh", title = "season 5", fontsize = 11)

plt.subplot(336)
grouped_df["season 6"].sort_values(ascending = True).plot(kind = "barh", title = "season 6", fontsize = 11)

plt.subplot(337)
grouped_df["season 7"].sort_values(ascending = True).plot(kind = "barh", title = "season 7", fontsize = 11)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
fig.suptitle("Trends in Screen Time Over 7 Seasons for Prominent Houses", fontsize=16)

plt.subplot(331)
grouped_df.loc["Stark"].plot(kind = 'line', title = "Stark", fontsize = 11, x_compat=True)

plt.subplot(332)
grouped_df.loc["Lannister"].plot(kind = 'line', title = "Lannister", fontsize = 11, x_compat=True)

plt.subplot(333)
grouped_df.loc["Targaryen"].plot(kind = 'line', title = "Targaryen", fontsize = 11, x_compat=True)

plt.subplot(334)
grouped_df.loc["Greyjoy"].plot(kind = 'line', title = "Greyjoy", fontsize = 11, x_compat=True)

plt.subplot(335)
grouped_df.loc["Tyrell"].plot(kind = 'line', title = "Tyrell", fontsize = 11, x_compat=True)

plt.subplot(336)
grouped_df.loc["Baratheon"].plot(kind = 'line', title = "Baratheon", fontsize = 11, x_compat=True)

plt.subplot(337)
grouped_df.loc["Martell"].plot(kind = 'line', title = "Martell", fontsize = 11, x_compat=True)

plt.subplot(338)
grouped_df.loc["Tully"].plot(kind = 'line', title = "Tully", fontsize = 11, x_compat=True)

plt.subplot(339)
grouped_df.loc["Arryn"].plot(kind = 'line', title = "Arryn", fontsize = 11, x_compat=True)

plt.show()
```

```python
ax = plt.axes()
ax.set_title("Heatmap showing the Variations in Screen Time of Prominent Houses for 7 Seasons", fontsize = 16)
sns.heatmap(grouped_df, annot=True, fmt = ".2f",  linewidth = 1, linecolor = 'k', robust = True, square = True, cmap = 'Blues', ax = ax)
```
